```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Ellipse
```

```python
plt.rc('xtick', labelsize=8)
plt.rc('ytick', labelsize=8)
plt.rc('axes', labelsize=8)
```

```python
np.random.seed(43)
```

```python
def k_reward(x,k=0.95):
    x = (x-0.1)/0.9
    y = (x-k*x)/(k-2*k*np.abs(x)+1)
    reward = 1-y
    return reward
```

```python
single_column_figure_width = 3.487
double_column_figure_width = 7

fig_width = single_column_figure_width
fig_height = fig_width / 1.618

fig, axs = plt.subplots(nrows=1,
                        ncols=1,
                        figsize=[fig_width,fig_height], # in inches
                        sharey=True)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=.05, hspace=.05)

axs.set_xlabel("Value for Objective 1")
axs.set_ylabel("Value for Objective 2")
axs.set_ylim([-0.25,1.25])
axs.set_xlim([0,1.5])

axs.set_xticklabels([])
axs.set_yticklabels([])

col = ["tab:blue","tab:orange","tab:purple"]
markerlist = ['P','*','v']
k = [0.5, 0, -0.9]
x_start = [0.1, 0.15, 0.13]
x_stop = [0.99, 0.85, 0.9]
y_offset = [0.1, -0.1, -0.12]

conformity_list = []
pfront_list = []

for i in range(3):
    conformity = np.linspace(x_start[i], x_stop[i], 100)
    pfront = k_reward(conformity,k[i]) + y_offset[i]
    
    conformity_list.append(conformity)
    pfront_list.append(pfront)

    axs.plot(conformity, pfront, linestyle='--', alpha=0.5, color=col[i])
    axs.scatter(conformity[0], pfront[0]  ,marker='s', color=col[i])
    axs.scatter(conformity[-1], pfront[-1]  ,marker='s', color=col[i])
    
    intrp_action1 = np.random.normal(loc = 0.3, scale = 0.05)
    intrp_action2 = np.random.normal(loc = 0.75, scale = 0.05)
    
    intrp_value1, intrp_value2 =  k_reward(np.array([intrp_action1, intrp_action2]), k[i]) + y_offset[i]
    axs.scatter(intrp_action1, intrp_value1, marker='P', color=col[i])
    axs.scatter(intrp_action2, intrp_value2, marker='*', color=col[i])

    
legend_elements_1 = [Line2D([0], [0], marker='s', 
                          color='w', label='Greedy Actions', 
                          markerfacecolor='k', markersize=7),
                  Line2D([0], [0], marker=markerlist[0], 
                          color='w', label='Sampled Action 1', 
                          markerfacecolor='k', markersize=9),
                  Line2D([0], [0], marker=markerlist[1], 
                          color='w', label='Sampled Action 2', 
                          markerfacecolor='k', markersize=10)]

legend_elements_2 = [
                  Line2D([0], [0], color=col[0], lw=1, linestyle='--', 
                         label=r'state $s_1$'),
                 Line2D([0], [0], color=col[1], lw=1, linestyle='--', 
                         label=r'state $s_2$'),
                  Line2D([0], [0], color=col[2], lw=1, linestyle='--', 
                         label=r'state $s_3$')]


ellipse1 = Ellipse(xy=(0.3,0.5), width=1.3, height=0.25, angle=92, alpha=0.2,
                        edgecolor=u'orchid', fc=u'orchid', lw=1.5,linestyle='--', )

ellipse2 = Ellipse(xy=(0.7,0.3), width=1.0, height=0.2, angle=90, alpha=0.2,
                        edgecolor=u'peru', fc=u'peru', lw=1.5,linestyle='--', )
axs.add_patch(ellipse1)
axs.add_patch(ellipse2)

leg1 = axs.legend(handles=legend_elements_1,
          loc="upper right",
           fontsize='x-small',
#            bbox_to_anchor=(-0.02,1.0,1.04,1),
#             mode="expand",
                 )

leg2 = axs.legend(handles=legend_elements_2,
          loc="lower right",
           fontsize='x-small',
#            bbox_to_anchor=(-0.02,1.0,1.04,1),
#             mode="expand",
                 )
axs.grid()

axs.add_artist(leg1)


fig.savefig('./figures/morl_inference.png', dpi=300, bbox_inches='tight', transparent=True)
```
